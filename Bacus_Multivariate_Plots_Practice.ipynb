```python
# import
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
fpath = '/content/drive/MyDrive/CodingDojo/01-Fundamentals/Week02/data-cleaning-core.csv'
```

```python
# inspect data
titvis_df = pd.read_csv(fpath)
titvis_df.info()
titvis_df.head()
```

```python
# plot the counts for the number of passengers by Sex, colored by if they Survived or not.
ax = sns.countplot(data = titvis_df, x ='Sex', hue = 'Survived');
ax.set(title = 'Male vs. Female Survivors');
```

- which group had a greater number of survivors (Survived=1)?

-female



---



```python
# Plot a barplot of the average number of "Parents or Children" for those that Survived vs. those that died
ax = sns.barplot(data = titvis_df, x='Survived', y='Parents or Children');
ax.set(title = 'Those that survived vs. Those that died');
```

- which group had a higher average number of Parents or Children?

-Survived has the higher average



---



```python
# plot a correlation heatmap of all of the numeric columns. Note: For this graph, you do not need an xlabel or ylabel. You only need a title
corr = titvis_df.corr(numeric_only = True)
corr.round(2)
fig, ax = plt.subplots(figsize=(12, 8))
sns.heatmap(corr, cmap = 'Greens', annot = True);
ax.set(title ='Correlation Greenmap')
```

- which column had the highest positive correlation to the "Survived" column?

- the Fare column
